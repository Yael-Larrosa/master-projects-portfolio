# En este notebook implementaremos el ELBOW-method

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist
```

```python
X, y_true = make_blobs(n_samples=300,cluster_std=0.6, random_state=0) # añadir aquí la definición del dataset

# Plotear los puntos
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
# Lista vacía donde guardaremos la pérdida en cada iteración
wcss = []

# Tarea: entrenamos i kmeans con inicialización 'k-means++' y random_state=0
# El bucle tiene 3 pasos:
#   1. La inicialización del moldeo
#   2. El entrenamiento
#   3. Añadir la WCSS a la lista 'wcss'. Hay un atributo del modelo que da ese valor
for i in range(1, 11):
    # Paso 1:
    kmeans = KMeans(n_clusters=i, random_state=0, init="k-means++")

    # Paso 2:
    kmeans.fit(X)
    # Paso 3:
    wcss.append(kmeans.inertia_)



# Ploteamos en un gráfica como varía la WCSS acorde con el número de clúster
plt.plot(range(1,11), wcss)
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
# En el eje vertical se debe apreciar el valor de WCSS y en el eje horizontal el número de clúster
# Recordad poner una etiqueta a los ejes para que sea más fácil de entender

```
