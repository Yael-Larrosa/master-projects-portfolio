```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs# Importar clase para crear clusters ficticios con sklearn

# Tarea: crear 3 clusters, con 2 características, la desviación del cluster que sea 0.5
X, y_true = make_blobs(n_samples=100, centers=3, n_features=2,cluster_std=0.5, random_state=0) # añadir aquí la definición del dataset

# Plotear los puntos con matplotlib
plt.scatter(X[:,0],X[:,1],c=y_true)
plt.show()
```

```python
# Entrenamos el algoritmo kmeans al dataset y obtenemos la predicción.
# Luego entenderemos como funciona

# 1. Importamos el modelo


# 2. Entrenamos el modelo

kmeans = KMeans(n_clusters=3, random_state=0, n_init="auto").fit(X)


# 3. Predecimos
prediccion= kmeans.predict(X, y_true)

print(prediccion)
```

```python
# Tarea: Obtener el "accuracy" del modelo.
from sklearn.metrics import accuracy_score
#como yo he generado los datos si que tengo y_true pero de normal no lo tendré por lo que no podré calcular el accuracy
score = accuracy_score(y_true, prediccion)
print(score)
```

En este caso, podemos sacar el accuracy. ¿Podremos en un dataset real de clustering?

```python
# Tarea: Obtenemos los centros de los clusters
centers = kmeans.cluster_centers_

print(centers)
```
